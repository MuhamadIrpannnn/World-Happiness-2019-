# STEP 6: TRANSFORM (DATA CLEANING)

# Mengganti nama kolom agar lebih sederhana dan konsisten

```python
df.rename(columns={
    'Overall rank': 'rank',
    'Country or region': 'country',
    'Score': 'happiness_score',
    'GDP per capita': 'gdp',
    'Social support': 'social_support',
    'Healthy life expectancy': 'life_expectancy',
    'Freedom to make life choices': 'freedom',
    'Generosity': 'generosity',
    'Perceptions of corruption': 'corruption'
}, inplace=True)
```

# Menampilkan hasil setelah rename

```python
df.head()
```

# Mengecek kembali missing value

```python
df.isnull().sum()
```

# Menghapus baris yang memiliki nilai kosong

```python
df = df.dropna()
```

# Mengecek jumlah data duplikat

```python
df.duplicated().sum()
```

# Menghapus data duplikat

```python
df = df.drop_duplicates()
```

____________________________________
# STEP 7: MEMBUAT DIMENSION TABLE

# Membuat tabel dimensi country (unik) dan Menambahkan primary key (ID) untuk setiap negara

```python
dim_country = df[['country']].drop_duplicates().reset_index(drop=True)
dim_country['country_id'] = dim_country.index + 1
```

_________________________________
# STEP 8: MEMBUAT DIMENSION YEAR

# Membuat tabel dimensi waktu (tahun)

```python
dim_year = pd.DataFrame({
    'year_id': [1],
    'year': [2019]
})
```

_______________________________
# STEP 9: MEMBUAT FACT TABLE

# Menggabungkan data utama dengan dimensi country, menambhakan foreign key year dan membuat fact table (hanya berisi data numerik/measure)

```python
fact = df.merge(dim_country, on='country')
fact['year_id'] = 1

fact_table = fact[[
    'country_id',
    'year_id',
    'happiness_score',
    'gdp',
    'social_support',
    'life_expectancy',
    'freedom',
    'generosity',
    'corruption'
]]
```

# Menampilkan hasil fact table

```python
fact_table.head()
```

_______________________________
# STEP 10: EXPORT DATA

# Menyimpan tabel dimensi dan fact ke file CSV

```python
dim_country.to_csv('dim_country.csv', index=False)
dim_year.to_csv('dim_year.csv', index=False)
fact_table.to_csv('fact_happiness.csv', index=False)
```
